```python
def batting(batters):
  batters['over_faced_first'] = batters['over_faced_first'].astype(float)
  batsman = batters.groupby(['batsman_id','series_name']).agg(
      matches=pd.NamedAgg(column='match id', aggfunc='count'),
      runs=pd.NamedAgg(column='runs', aggfunc='sum'),
      balls_faced=pd.NamedAgg(column='balls_faced', aggfunc='sum'),
      Fours=pd.NamedAgg(column='Fours', aggfunc='sum'),
      Sixes=pd.NamedAgg(column='Sixes', aggfunc='sum'),
      fifties=pd.NamedAgg(column='runs', aggfunc=lambda x: (x > 50).sum()),
      thirties=pd.NamedAgg(column='runs', aggfunc=lambda x: (x > 30).sum()),
      high_sr=pd.NamedAgg(column='strike_rate', aggfunc=lambda x: (x > 140).sum()),
      over_faced_first=pd.NamedAgg(column='over_faced_first', aggfunc='mean'),
      captaincy=pd.NamedAgg(column='is_batsman_captain', aggfunc='sum'),
      wins=pd.NamedAgg(column='won', aggfunc='sum')
  )
  batsman['avg_sr'] = batsman['runs']*100/batsman['balls_faced']
  batsman['boundaries'] = batsman['Fours']+batsman['Sixes']
  batsman.drop(['Fours','Sixes'],axis=1,inplace=True)
  return batsman
batsman_details = batting(bat)
batsman_details
```
